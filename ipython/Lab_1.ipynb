## Data Collection

* For a given trial, start the cart at the top of the ramp, hit record, and release the cart from rest. Stop recording once the cart reaches the bottom. Scale your data appropriately. 
* To determine the acceleration, we will be taking the slope of the velocity plot; recall $a = \frac{dv}{dt}$
<br>
To calculate the slope, first use the select tool to select the relevant data points in the bounding box (only select the points corresponding to motion down the ramp). <img src="https://github.com/phelps-matthew/Studio-Physics-Labs/raw/master/images/common/select.png" width="40" border="1">
<br>
Next use the slope tool and select <i>Linear: mx+b</i>. The value of $m$ is the acceleration. <img src="https://github.com/phelps-matthew/Studio-Physics-Labs/raw/master/images/common/slope.png" width="30" border="1">
* Record the acceleration of the cart for the following six values of $\theta : \{10, 15, 20, 25, 30, 35\}$. Store the acceleratoin measurements into the list `accels` below. 
* The list should follow the format `np.array([a1, a2, a3, .., a6])`

```
thetas = np.arange(10, 40, 5)
accels = np.array([0.932, 1.75, 2.67, 3.39, 4.19, 4.88])
```

## Plotting Experimental Data

* To get a feel for your data, let's plot it. We will include appropriate axis labels, i.e. variable and units (for example "a (m/s^2)" ), and make the plot points red.

```
sns.set()
axes_0 = sns.scatterplot(thetas, accels, color='red')
axes_0.set(xlim=0, ylim=0, xlabel="theta (deg)", ylabel="a (m/s^2)")
plt.show()
```

### Linear Fitting

Given our experimental data, we may ask, Can we approximate $a(\theta)$ as a linear relationship? If so, what is the equation of the line that describes this relationship?

To determine this, we assume $a(\theta)$ to be of the form $a=m\theta = b$ and perform a linear regression to find $m$ and $b$. 

* Use the code below to find the slope ($m$) and intercept ($b$) that minimizes the squared distance between each data point and the line. This is the line of best fit. You will use the linear regression parameters within the analysis section. 

```
slope, intercept, r_value, p_value, std_err = stats.linregress(thetas, accels)
ax = sns.regplot(thetas, accels, scatter_kws=dict(color="red"))
ax.set(xlim=(0,40), ylim=0, xlabel="theta (deg)", ylabel="a (m/s^2)")
plt.show()
print("slope={:.3f}\nintercept={:.3f}\nr_value={:.3f}".format(slope, intercept, r_value))
```
